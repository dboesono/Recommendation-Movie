## Built Data Preprocessing Function

```python
def preprocess_data(df):
    """Preprocess data into training, testing, and validation dataset"""
    # df = df.sample(frac=0.001)
    # Cut the dataframe
    top_10_percent = int(0.001 * len(df))
    df = df.head(top_10_percent)
    
    df = df[['Movie_ID', 'User_ID', 'Rating']] # Change column order
    train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42) # Split training set
    val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42) # Split testing and validation set
    
    # Sort datasets by their Movie_ID
    dataframes = [train_df, val_df, test_df]
    for i in range(len(dataframes)):
        dataframes[i] = dataframes[i].sort_values(by='Movie_ID')
        
    return dataframes # Output datasets
```

## Preprocess Dataset

```python
df = pd.read_csv('../data/raw/Netflix_Dataset_Rating.csv')
df = df.sort_values(by='Movie_ID')[['Movie_ID', 'User_ID', 'Rating']]
df
```

```python
train_df, val_df, test_df = preprocess_data(df)
```
